```python
# Very nice, there are no duplicated rows in the dataframe! Now let's check and see if there is any missing data
n_records = len(sa_women)
def missing_values_df(df):
    for column in df:
        print("{} | {} | {}".format(
            column, len(df[df[column].isnull()]) / (1.0*n_records), df[column].dtype
        ))

missing_values_df(sa_women)
```

```python
## Great, there is no missing data! Now, let's check for outliers
from scipy import stats
import numpy as np

# we can take a sample column to see its outliers
column = "total_households"
print(f"the average for column {column} is {sa_women[column].mean()}")
#we consider outliers those observations with a z score (absolute) greater than 3
sa_women[(np.abs(stats.zscore(sa_women[column])) > 3)].head()
```

```python
def outliers_col(df):
    for column in df:
        if df[column].dtype != np.object:
            n_outliers = len(df[(np.abs(stats.zscore(df[column])) > 3)& \
                  (df[column].notnull())
                 ])
            print("{} | {} | {}".format(
                df[column].name,
                n_outliers,
                df[column].dtype
        ))

outliers_col(sa_women)
```

```python
## Let's check out all of the numeric variable distributions with an interactive widget

from ipywidgets import interact, fixed

def numerical_variable_distribution(df, col):
    df[col].plot.kde()
```

```python
numerical_columns = sa_women.select_dtypes(include=[np.number]).columns.tolist()
interact(numerical_variable_distribution, 
         col=numerical_columns, df=fixed(sa_women));
```

```python
## We see that a lot of these distributions are skewed, especially in the columns with a high number of outliers,
# as previously listed. Let's see how these columns compare to perfect normality, and see if any can be considered
# normal

from scipy import stats

def numerical_variable_normality(col):
    stats.probplot(sa_women[col], plot=sns.mpl.pyplot) #Note we can access our plot environment using sns.mpl.pyplot
```

```python
interact(numerical_variable_normality, col=numerical_columns);
```

```python
## It seems that our variables fail to follow the perfectly normal plot, but let's see if, mathematically,
## they could be considered normal by doing a normal test, and if they pass, they will be listed

for num_col in numerical_columns:
    _, pval = stats.normaltest(sa_women[num_col][sa_women[num_col].notnull()])
    if(pval >= 0.05):
        print("Column {} appears to follow a normal distribution".format(num_col))
```

```python
## None of our numeric variables are normally distributed, it seems, which could pose problems with some estimators.
## This will have to be resolved when making our models to forecast the target, as some algorithms relies on the 
## assumption of normally distributed independent variables.

## Now, let's look for independent variables (in their current form) that are correlated with our target variable.

sa_women.corr()['target'][abs(sa_women.corr()['target']) > 0.5].sort_values()
```
